```python
wine = pd.read_csv(filepath_or_buffer=r"C:\Users\santo\desktop\wine.csv", sep=",", encoding="latin1")    
```

```python
#take a random sample
random_subset = wine.sample(n=100)
```

```python
result = random_subset.shape
print("NUMBER OF ROWS AND COLUMNS:")
print(result)
print()
```

```python
print("FILE INFORMATION:")
random_subset.info()
print()
```

```python
result = random_subset.describe()
print("SUMMARY DESCRIPTIVE STATS FOR NUMERICAL COLUMNS:")
print(result)
print()
```

```python
result = random_subset["country"].value_counts()
print("FREQUENCY DISTRIBUTION FOR CATEGORICAL COLUMN COUNTRY:")
print(result)
print()
```

```python
result = random_subset["region_1"].value_counts()
print("FREQUENCY DISTRIBUTION FOR CATEGORICAL COLUMN REGION_1:")
print(result)
print()
```

```python
result = random_subset.apply(lambda x: sum(x.isnull()), axis=0) 
print("MISSING VALUES (NAN) BY COLUMNS:")
print(result)
print()
```

```python
#replace missing values with the median
random_subset['price'].fillna(random_subset['price'].median(), inplace=True)
```

```python
#check that there are no more missing values in price
result = random_subset.apply(lambda x: sum(x.isnull()), axis=0) 
print("MISSING VALUES (NAN) BY COLUMNS:")
print(result)
print()
```

```python
result = random_subset.duplicated()
print("DUPLICATED ROWS:")
print(result)       
print()
```

```python
random_subset = random_subset.drop_duplicates(keep="first")
print("DROP DUPLICATED ROWS AND KEEP THE FIRST:")
print(random_subset)       
print()
```

```python
random_subset.to_csv(r"C:\Users\santo\desktop\wine_output_data.csv", encoding="latin1")   
print("THE OUTPUT FILE HAD BEEN CREATED SUCCESSFULLY:")
print("wine_output_data.csv")
print()   
```
